```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('abuse_address_new.csv',header = None)
df.columns = ['address', 'category', 'price']
df.head(2)
```

```python
hist_df = pd.concat([df['category'], df['price']], axis=1)
key = ['sextortion', 'scam']

for k in key:
    check = hist_df['category'] == str(k)
    check_df = hist_df[check]
    xticks = [i for i in range(0, 11000, 500)]
    plt.figure(figsize=(30,10))
    plt.title(str(k))
    plt.hist(check_df['price'], range = [0, 10000], bins = 100)
    plt.ylim(0, 400)
    plt.xticks(xticks)
    plt.show()
```
