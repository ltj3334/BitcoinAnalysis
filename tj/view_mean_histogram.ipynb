```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
files = []
for file in os.listdir('./'):
    if "mean_std_result" in file:
        files.append(file)
```

```python
for file in files:
    data = pd.read_csv(file, header=None)
    data.columns = ['addr', 'mean', 'std']
    plt.title(f"{file}")
    plt.hist(data['mean'], range=(0, 1000), bins=100)
    plt.show()
```
