```python
from sklearn.model_selection import train_test_split

y = data['label']
x = data.drop(['label','wallet_id'], axis = 1)
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, stratify=y, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score # 정확도 함수

rf_clf = RandomForestClassifier(n_estimators=20, max_depth=5,random_state=0)
rf_clf.fit(train_x,train_y)

rf_predict = rf_clf.predict(test_x)
print(f"Random Forest Accuracy : {accuracy_score(test_y,rf_predict)}")


lr_clf = LogisticRegression(C=154.3669, penalty="l1",solver='liblinear',tol=1)
lr_clf.fit(train_x,train_y)
lr_predict = lr_clf.predict(test_x)
print(f"Logistic Regression Accuracy : {accuracy_score(test_y,lr_predict)}")

```
